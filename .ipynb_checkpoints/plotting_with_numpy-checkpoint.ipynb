# Plotting with Numpy and Matplolib

the main command in matplotlib :
plt.plot(numpy array, numpy array of same length)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# plot a straight line 
plt.plot(np.arange(10), np.arange(10))
```

```python
# plot an exponential
plt.plot(np.arange(10), np.exp(np.arange(10)))
```

```python
# place two graphs side by side
plt.figure()
plt.subplot(121)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(122)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

```python
# place two graphs top to bottom
plt.figure()
plt.subplot(211)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(212)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

```python
# the first two numbers in subplot command refer to grid size
# the last digit refers to traversing the grid in reading order
# left to right and then up to down
plt.figure()
plt.subplot(311)
plt.subplot(312)
plt.subplot(313)
```

```python
# plot in a 2x2 grid traversing in a reading order 1 -4 
plt.figure()
plt.subplot(221)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(222)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.subplot(223)
plt.plot(np.arange(10), np.arange(10))
plt.subplot(224)
plt.scatter(np.arange(10), np.arange(10))
```
